```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

data = {
    "X": [1, 2, 3, 4, 5],
    "y": [2, 4, 6, 7, 10]
} 

df=pd.DataFrame(data)


X = df[['X']] #2d dataframe 
print(X,'\n')

y = df['y'] #1d dataframe
print(y,'\n')

model = LinearRegression()
model.fit(X, y)
#y = w0+w1*x

print('\n',model.coef_[0],'\n') #slope or tangen
print('\n',model.intercept_,'\n') #intercept

# t = model.coef_[0]
# i = model.intercept_
value = 7

# print('\n',i+t*value,'\n')

g = model.predict([[value]])
print(g)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split as ts
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
import numpy as np

sns.set_style('whitegrid')
df =  pd.read_csv('advertising.csv')


df.head()
# sns.pairplot(df)
# plt.show()
print(df)

x = df[['TV']]
print(x)

y= df[['Sales']]
print(y)

x_train, x_test, y_train, y_test = ts(x, y, test_size=0.2, random_state=42)

print('train_set_shape',x_train.shape,'\n')
print('train_set_shape',x_test.shape,"\n")

model = LinearRegression()
model.fit(x_train, y_train)


print(model.coef_[0])  
print(model.intercept_)

print("\n",model.predict([[10]]))  

plt.figure(figsize=(8, 6))
plt.scatter(x_test, y_test, color='blue', label='Actual Sales')
plt.plot(x_test, model.predict(x_test), color='red', linewidth=2, label='Predicted Sales (Regression Line)')
plt.title('TV Ad Spending vs. Sales (Test Set)')
plt.xlabel('TV Ad Spending ($ thousands)')
plt.ylabel('Sales (thousands of units)')
plt.legend()
plt.show()

y_pred_simple = model.predict(x_test)
print("\n",y_pred_simple)

mse_value = metrics.mean_squared_error(y_test, y_pred_simple)
print("\n",mse_value)
print('\n',type(mse_value))

rmse_value = np.sqrt(mse_value)
print("\n",'root_mean_squared_error',rmse_value)
```
