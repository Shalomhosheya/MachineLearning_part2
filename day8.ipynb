```python
#linear classification(binary) - logistic regression
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
#filter no of awuful words and awesome words from dataset
from sklearn.feature_extraction.text import CountVectorizer
#calculate the model accuracy
from sklearn.metrics import accuracy_score
import joblib as joblib


df=pd.read_csv('womens_clothing_ecommerce_reviews.csv')
print(df.head())
df.info()

x=df['Review Text']
y=df['sentiment']

#'stratify=y ensure the propotion of postive and nagative reviews
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)

#stop english words like 'is','the', which don't carry much sentimate
vectorizer = CountVectorizer(stop_words='english')
#fit the vectorizer on the training data and transform it(wachana tika aragena vocabulary ekak hadagena eke count eka gannawa.)
X_train_vectorized = vectorizer.fit_transform(X_train)
#transform the test data
X_test_vectorized = vectorizer.transform(X_test)
# Create and train the logistic regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vectorized, Y_train)
print("model trained successfully")





# Make predictions on the test data
y_pred = model.predict(X_test_vectorized)
# Calculate the accuracy of the model
accuracy = accuracy_score(Y_test, y_pred)
print("Model Accuracy:", accuracy)

#save model
joblib.dump(model, 'model.joblib')

# Display the coefficients of the model
coefficients = model.coef_[0]
print("Model Coefficients:", coefficients)

#add customize reviews and predict its sentiment
custom_review = ["This product is amazing!", "I hate this item, it's terrible.", "It's okay, not great but not bad either.",
                    "Absolutely love it!", "Worst purchase ever.", "It's just fine, nothing special."]
custom_review_vectorized = vectorizer.transform(custom_review)
custom_predictions = model.predict(custom_review_vectorized)
for review,sentiment in zip(custom_review, custom_predictions):
    print(f"Review: {review} | Predicted Sentiment: {sentiment}")
```

```python
#linear classification(binary) - logistic regression
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
#filter no of awuful words and awesome words from dataset
from sklearn.feature_extraction.text import CountVectorizer
#calculate the model accuracy
from sklearn.metrics import accuracy_score
import joblib as joblib

from sklearn.pipeline import Pipeline


df=pd.read_csv('womens_clothing_ecommerce_reviews.csv')
print(df.head())
df.info()

x=df['Review Text']
y=df['sentiment']

#'stratify=y ensure the propotion of postive and nagative reviews
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)


#define the pipeline
pipeline=Pipeline([
    ('vectorizer',CountVectorizer(stop_words='english')),
    ('classifier',LogisticRegression(max_iter=1000))
])
#train entire pipline
print("training the model using pipeline...")
pipeline.fit(X_train,Y_train)
print("pipeline trained successfully...!")
```
